**Ques 1.1** Write a Python Program(with class concepts) to find the area of the triangle using the below formula.

Area = (s(s-a)*(s-b)*(s-c)) ** 0.5

```python
class Sides:
    def __init__(self, a, b, c):
        self.a = float(a)
        self.b = float(b)
        self.c = float(c)
        self.area = 0

class Area(Sides):
    def __init__(self, a, b, c):
        Sides.__init__(self, a, b, c)

    def set_area(self):
        s = (self.a + self.b + self.c) / 2
        self.area = float((s * (s - self.a) * (s - self.b) * (s - self.c))) ** 0.5

    def print_area(self):
        return self.area     
```

```python
a, b, c = input("a = "), input("b = "), input("c = ")
obj = Area(a, b, c)
obj.set_area()
print(f"Area : {obj.print_area()}")
```

**Ques 1.2** Write a function filter_long_words() that takes a list of words and an integer n and returns
the list of words that are longer than n.

```python
def filter_long_words(lis, n):
    return [_ for _ in lis if len(_)>n]
```

```python
filter_long_words(["apple", "banana", "mango", "orange", "kiwi"], 4)
```

**Ques 2.2** Write a Python function which takes a character (i.e. a string of length 1) and returns True if
it is a vowel, False otherwise.

```python
def isVowel(elt):
    ls = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
    return elt in ls
```

```python
isVowel('Q')
```
