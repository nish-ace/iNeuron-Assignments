**Ques1.1** Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()

```python
def func1(a, b):
  return a+b
```

```python
def myreduce(func, ls):
  sum= ls[0]
  for i in range(1,len(ls)):
    temp=ls[i]
    sum=func1(sum, temp)
  return sum
```

```python
myreduce(func1, [10, 20, 30, 40])
```

**Ques1.2** Write a Python program to implement your own myfilter() function which works exactly
like Python's built-in function filter()

```python
def func2(var):
  if type(var)==int:
    return True
  return False
```

```python
def myfilter(func2, param):
  for _ in param:
    if func2(_):
      print(_)
```

```python
myfilter(func2, ('A', 1, 'B', 2))
```

**Ques2.** Write List comprehensions to produce the following Lists:

(b)['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']

```python
str = "xyz"
ls = [char*num for char in str for num in range(1,5)]
ls
```
