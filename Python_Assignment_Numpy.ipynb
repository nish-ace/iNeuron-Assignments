**Ques 1** Write a function so that the columns of the output matrix are powers of the input vector.

```python
def Power(vect, col, increasing):
    if increasing is True:
        arr = np.array([pow(x, i) for x in vect for i in range(col)])

    if increasing is False:
        arr = np.array([pow(x, len(vect)-i-1) for x in vect for i in range(col)])
    
    return arr.reshape(len(vect), col)
```

```python
Power([1, 2, 3, 4], 4, True)
```

```python
Power([1, 2, 3, 4], 4, False)
```

**Ques 2** Write a function to find moving average in an array over a window: 

Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.

```python
def movingAvg(ls, k):
    arr = np.cumsum(ls, dtype=int)
    arr[k:] = arr[k:] - arr[:-k]
    return arr[k-1:]/k
```

```python
movingAvg([3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150], 3)
```
