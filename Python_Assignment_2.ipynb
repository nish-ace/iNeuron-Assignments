**Ques1.** Create the below pattern using nested for loop in Python.

```python
for i in range(0, 5):
  for j in range(0, i):
    print("*", end="")
  print(end='\n')

for i in range(5, 0 , -1):
  for j in range(0, i):
    print("*", end="")
  print(end='\n')
```
